# Importing Libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# Importing Dataset

```python
dataset = pd.read_csv('sonar_data.csv')
```

```python
print("Number of rows and columns : ",dataset.shape)
```

```python
dataset.head()
```

### Statistical measures of the dataset

```python
dataset.describe()
```

### Counting number of rocks and number of mines according to dataset

```python
dataset['R'].value_counts()
```

* M ----> Mine
* R ----> Rock

```python
dataset.groupby("R").mean()
```

### Separating data and labels

```python
X = dataset.drop(columns = 'R', axis = 1)
Y = dataset["R"]
```

```python
X
```

```python
Y
```

# Splitting the data into Training data and Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, stratify=Y, random_state=1)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

# Model Training

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

# Model Evaluation

### Accuracy on the training data

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy on training data : ", training_data_accuracy)
```

### Accuracy on the testing data

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy on testing data : ", test_data_accuracy)
```
